```python
df.columns = [str(c).strip() for c in df.columns]
if 'InvoiceDate' in df.columns:
    df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
# Numeric columns
for col in ['Quantity','UnitPrice']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
# IDs to string
for col in ['CustomerID','InvoiceNo']:
    if col in df.columns:
        df[col] = df[col].astype('string').str.strip()
# Create TotalPrice
if 'Quantity' in df.columns and 'UnitPrice' in df.columns:
    df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df[['InvoiceDate','Quantity','UnitPrice','CustomerID','InvoiceNo','TotalPrice']].head()
```

```python
# Feature engineering and basic cleaning
# Time features
if 'InvoiceDate' in df.columns:
    df['Invoice_Year'] = df['InvoiceDate'].dt.year
    df['Invoice_Month'] = df['InvoiceDate'].dt.month
    df['Invoice_Day'] = df['InvoiceDate'].dt.day
    df['Invoice_Hour'] = df['InvoiceDate'].dt.hour
# Duplicates and negative quantities (often cancellations are marked by 'C' in InvoiceNo)
print('duplicates:', df.duplicated().sum())
if 'Quantity' in df.columns:
    print('negative quantities count:', (df['Quantity'] < 0).sum())
```

```python
# Summary statistics and simple outlier detection (IQR)
print(df.describe(include='all').T)
for col in ['Quantity','UnitPrice','TotalPrice']:
    if col in df.columns:
        ser = df[col].dropna()
        q1 = ser.quantile(0.25)
        q3 = ser.quantile(0.75)
        iqr = q3 - q1
        lower = q1 - 1.5*iqr
        upper = q3 + 1.5*iqr
        print(f'\n{col}: q1={q1}, q3={q3}, iqr={iqr}, lower={lower}, upper={upper}, outliers={(ser<lower).sum() + (ser>upper).sum()}')
```

```python
# Plots: histogram of TotalPrice and time-series counts by month
import matplotlib.pyplot as plt
if 'TotalPrice' in df.columns:
    plt.figure(figsize=(8,4))
    plt.hist(df['TotalPrice'].dropna(), bins=60)
    plt.title('Distribution of TotalPrice')
    plt.xlabel('TotalPrice')
    plt.ylabel('Frequency')
    plt.show()

if 'InvoiceDate' in df.columns:
    by_month = df.set_index('InvoiceDate').resample('M').size()
    plt.figure(figsize=(10,4))
    plt.plot(by_month.index, by_month.values)
    plt.title('Transactions per Month')
    plt.xlabel('Month')
    plt.ylabel('Count')
    plt.show()
```


# Customer Segmentation using K-Means

This notebook performs:
- Cluster visualization
- Assigning cluster labels to original data
- Segment-wise statistical analysis
- Customer profiling using visualizations


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

## Load Dataset

```python

# Update path if required
df=pd.read_excel(r"C:\Users\MANISH SHUKLA\Downloads\Online Retail.xlsx")
df.head()
```

## Data Preprocessing

```python

# Example preprocessing: remove missing values and select numeric features
df_clean = df.dropna()

numeric_features = df_clean.select_dtypes(include=np.number)
numeric_features.head()
```

## Standardize the Data

```python

scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_features)
```

## Apply K-Means Clustering

```python

kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

df_clean['Cluster'] = clusters
df_clean.head()
```

## Visualize Clusters

```python

plt.figure()
plt.scatter(
    scaled_data[:, 0],
    scaled_data[:, 1],
    c=clusters
)
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    marker='X'
)
plt.title("Customer Segments")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## Cluster-wise Summary Statistics

```python

num_cols = df_clean.select_dtypes(include="number").columns
cluster_summary = df_clean.groupby("Cluster")[num_cols].agg(["mean", "median", "std"])
cluster_summary
```

## Customer Segment Profiling

```python

for col in numeric_features.columns:
    plt.figure()
    sns.boxplot(x='Cluster', y=col, data=df_clean)
    plt.title(f'{col} by Customer Segment')
    plt.show()
```


## Segment Interpretation

Based on the statistics and visualizations:
- Each cluster represents a distinct customer segment
- Differences can be seen in purchasing behavior and value
- These profiles can be used for targeted marketing and business strategy
